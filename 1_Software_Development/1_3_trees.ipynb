```python
# As a Software Architect and a Security Expert and a Site Reliability Engineering can you update this code above to do the following and check and update all the functions above
# 1. Add type hints for better readability and maintainability.
# 2. Add docstrings to classes and methods for better documentation.
# 3. Implement input validation to ensure only valid keys are inserted.

import sys
from typing import Optional, Tuple
from threading import Lock

class BSTNode:
    def __init__(self, data: int) -> None:
        """
        Initialize a new node with the given data.
        
        :param data: The data to be stored in the node.
        """
        self.data: int = data
        self.left: Optional['BSTNode'] = None
        self.right: Optional['BSTNode'] = None

class BinarySearchTree:
    def __init__(self) -> None:
        """
        Initialize an empty Binary Search Tree.
        """
        self.root: Optional[BSTNode] = None
        self.size: int = 0
        self.lock = Lock()

    def insert(self, data: int) -> None:
        """
        Insert a new node with the given data into the BST.
        
        :param data: The data to be inserted.
        :raises ValueError: If the data is not an integer.
        """
        if not isinstance(data, int):
            raise ValueError("Only integer values are allowed")
        
        with self.lock:
            if self.root is None:
                self.root = BSTNode(data)
            else:
                self._insert(self.root, data)
            self.size += 1

    def _insert(self, node: BSTNode, data: int) -> None:
        """
        Helper method to insert a new node into the BST.
        
        :param node: The current node in the BST.
        :param data: The data to be inserted.
        """
        if data < node.data:
            if node.left is None:
                node.left = BSTNode(data)
            else:
                self._insert(node.left, data)
        else:
            if node.right is None:
                node.right = BSTNode(data)
            else:
                self._insert(node.right, data)

    def print_tree(self) -> None:
        """
        Print the elements of the BST in order.
        """
        with self.lock:
            if self.root is not None:
                self._print_tree(self.root)
                print()  # for a new line after printing the tree

    def _print_tree(self, node: Optional[BSTNode]) -> None:
        """
        Helper method to print the elements of the BST in order.
        
        :param node: The current node in the BST.
        """
        if node is not None:
            self._print_tree(node.left)
            print(str(node.data), end=' ')
            self._print_tree(node.right)

    def search(self, data: int) -> Optional[BSTNode]:
        """
        Search for a node with the given data in the BST.
        
        :param data: The data to search for.
        :return: The node with the given data, or None if not found.
        """
        with self.lock:
            return self._search(self.root, data)

    def _search(self, node: Optional[BSTNode], data: int) -> Optional[BSTNode]:
        """
        Helper method to search for a node with the given data in the BST.
        
        :param node: The current node in the BST.
        :param data: The data to search for.
        :return: The node with the given data, or None if not found.
        """
        if node is None or node.data == data:
            return node
        if data < node.data:
            return self._search(node.left, data)
        return self._search(node.right, data)

    def delete(self, data: int) -> None:
        """
        Delete a node with the given data from the BST.
        
        :param data: The data to delete.
        """
        with self.lock:
            self.root, deleted = self._delete(self.root, data)
            if deleted:
                self.size -= 1

    def _delete(self, node: Optional[BSTNode], data: int) -> Tuple[Optional[BSTNode], bool]:
        """
        Helper method to delete a node with the given data from the BST.
        
        :param node: The current node in the BST.
        :param data: The data to delete.
        :return: A tuple containing the new root of the subtree and a boolean indicating if the node was deleted.
        """
        if node is None:
            return node, False

        if data < node.data:
            node.left, deleted = self._delete(node.left, data)
        elif data > node.data:
            node.right, deleted = self._delete(node.right, data)
        else:
            if node.left is None:
                return node.right, True
            elif node.right is None:
                return node.left, True

            temp = self._min_value_node(node.right)
            node.data = temp.data
            node.right, _ = self._delete(node.right, temp.data)
            return node, True

        return node, deleted

    def _min_value_node(self, node: BSTNode) -> BSTNode:
        """
        Helper method to find the node with the minimum value in the BST.
        
        :param node: The current node in the BST.
        :return: The node with the minimum value.
        """
        current = node
        while current.left is not None:
            current = current.left
        return current

    def get_size(self) -> int:
        """
        Get the number of nodes in the BST.
        
        :return: The number of nodes in the BST.
        """
        with self.lock:
            return self.size

    def get_memory_usage(self) -> int:
        """
        Get the total memory usage of the BST.
        
        :return: The total memory usage of the BST in bytes.
        """
        with self.lock:
            return sys.getsizeof(self) + self._get_memory_usage(self.root)

    def _get_memory_usage(self, node: Optional[BSTNode]) -> int:
        """
        Helper method to calculate the memory usage of the BST.
        
        :param node: The current node in the BST.
        :return: The memory usage of the BST in bytes.
        """
        if node is None:
            return 0
        return sys.getsizeof(node) + self._get_memory_usage(node.left) + self._get_memory_usage(node.right)

# Example usage
bst = BinarySearchTree()
bst.insert(10)
bst.insert(5)
bst.insert(15)
bst.insert(3)
bst.insert(7)
bst.insert(12)
bst.insert(18)

print("Tree elements in order:")
bst.print_tree()

print("Search for 7:", bst.search(7) is not None)
print("Search for 20:", bst.search(20) is not None)

bst.delete(10)
print("Tree elements after deleting 10:")
bst.print_tree()

print("Number of active items:", bst.get_size())
print("Total memory usage:", bst.get_memory_usage(), "bytes")
```
