```python
from collections import Counter
import re

#  Please write plain Python code  using the base language to download text from a given UTL and then count every instance of every word in that text.

import urllib.request

# Function to download text from a given URL
def download_text(url):
    response = urllib.request.urlopen(url)
    text = response.read().decode('utf-8')
    return text

# Function to count word instances in the text
def count_words(text):
    words = re.findall(r'\b\w+\b', text.lower())
    word_count = Counter(words)
    return word_count

# Example usage
url = 'https://www.deeplearning.ai/the-batch/issue-270/'
text = download_text(url)
word_count = count_words(text)

print(word_count)
```
