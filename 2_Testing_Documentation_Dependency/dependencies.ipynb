```python
# Can you create a 3 dimensional numpy array with a shape of (3, 3, 3) and have random values in it?
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Create a 3-dimensional numpy array with random values
array_3d = np.random.rand(3, 3, 3)
print(array_3d)

# Now can you create a 3 dimensional visualisation of the array using matplotlib?
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Get the coordinates of the points in the array
x, y, z = np.indices(array_3d.shape)

# Plot the points
ax.scatter(x, y, z, c=array_3d.flatten(), cmap='viridis')

plt.show()

```
